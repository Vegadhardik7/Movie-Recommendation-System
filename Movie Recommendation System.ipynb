### Formating Data

```python
# This function is use to convert string to list
import ast

# Example:
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def getGenres(val):
    lst = []
    for i in ast.literal_eval(val):
        lst.append(i['name'])
    return lst
```

```python
data['genres'] = data['genres'].apply(getGenres)
data.head(1)
```

```python
data['keywords'] = data['keywords'].apply(getGenres)
```

```python
data.head(1)
```

```python
def getConvert(val):
    lst = []
    cnt = 1
    for i in ast.literal_eval(val):
        if cnt!=3:
            lst.append(i['name'])
            cnt+=1
        else:
            break
    return lst

data['cast'] = data['cast'].apply(getConvert)
```

```python
data.head(1)
```

```python
def getJobdir(val):
    lst = []
    for i in ast.literal_eval(val):
        if i['job'] == 'Director':
            lst.append(i['name'])
            break
    return lst

data['crew'] = data['crew'].apply(getJobdir)
```

```python
data.head(1)
```

```python
# converting overview from string to list

data['overview'] = data['overview'].apply(lambda x:x.split())
```

```python
data.head(1)
```

```python
# Removing Spaces form genres, keywords, cast, crew

data['genres'] = data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords'] = data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast'] = data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew'] = data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head()
```

```python
# Tags : Concatination of all the above columns

data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
data.head()
```

```python
new_df = data[['id', 'title', 'tags']]
new_df.head()
```

#### Convet list 'tags' to string

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

### Vectorizing Process

```python
# actor and actors are some so we have to deal with it (So We Are Applying Stemmping)

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names() # Most occuring 5000 words
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
# Top 5 movies
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
